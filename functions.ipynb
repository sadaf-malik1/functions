  # Iterables
An iterable is any Python object capable of returning its members one at a time, allowing it to be looped over in a for-loop. Examples include lists, tuples, dictionaries, sets, and even strings. Basically, if an object can return an iterator when used with the iter() function, it's considered an iterable.

 # Iterators
An iterator is an object that represents a stream of data. It performs the actual iteration over the iterable container and returns data one element at a time. In Python, an iterator object must implement two methods: __iter__() and __next__().

Here's a simple example to illustrate the difference:

Iterable Example:
python

Key Differences
Creation:

Iterable: Can be directly used in a for loop or with other constructs that expect an iterable.

Iterator: Obtained from an iterable using the iter() function.

Usage:

Iterable: Holds data and can produce an iterator.

Iterator: Traverses through the elements of the iterable one by one using __next__().

Methods:

Iterable: Implements __iter__().

Iterator: Implements both __iter__() and __next__().

Reusability:

Iterable: Can be iterated multiple times (e.g., lists, tuples).

Iterator: Can be iterated only once, because it maintains state and its elements get consumed.

  Generators in Python are a special type of iterator that allows you to iterate over a sequence of values in a memory-efficient manner. They are defined using the yield keyword and are ideal for working with large datasets or streams of data, as they generate values on the fly and do not store the entire sequence in memory.

# Defining a Generator
Generators are defined similarly to regular functions but use the yield keyword instead of return. Each time the yield statement is executed, the function's state is "saved" and can be resumed later.

Here's a simple example to illustrate:

python
def simple_generator():
    yield 1
    yield 2
    yield 3

 Creating a generator object
gen = simple_generator()

 Iterating through the generator
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3


Q7 What are the advantages of using generators over regular function ?

  Using generators offers several advantages over regular functions, especially when it comes to memory management and performance are:

# Memory Efficiency:

Generators generate values on the fly and do not store the entire sequence in memory. This makes them ideal for working with large datasets or streams of data, as they only consume memory for the current value.

# Lazy Evaluation:

Generators produce values only when requested, allowing you to handle potentially infinite sequences or large files without running out of memory.

# Improved Performance:

Generators can significantly improve performance by avoiding the overhead of creating and storing large data structures. They yield values one at a time, reducing the time required to create and manage data.

# Simpler Code:

Generators can simplify code by eliminating the need for explicit iterator management. They provide a clean and concise way to define iterators using the yield keyword, making the code more readable and maintainable.

# State Persistence:

Generators maintain their state between iterations, allowing you to create complex iteration logic without the need for additional variables or structures. Each time yield is called, the generator's state is saved, and it resumes from where it left off on the next iteration.

Example: Reading a Large File
Here's an example of using a generator to read a large file line by line, without loading the entire file into memory:


def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

 Using the generator to read a file

Q10 What is the difference between 'map()' , 'reduce()' ,and 'filter()' functions in python ?

 The map(), reduce(), and filter() functions are all built-in functions in Python that are commonly used in functional programming. While they have similarities in that they process iterables, they serve different purposes and have distinct behaviors:

# 1. map()
The map() function applies a given function to each item in an iterable (e.g., list, tuple) and returns an iterator of the results.

Purpose: To transform or apply a function to each element in an iterable.

Syntax: map(function, iterable)

Example:

python
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x**2, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]
# 2. reduce()
The reduce() function applies a given function cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value. reduce() is part of the functools module, so it needs to be imported before use.

Purpose: To reduce an iterable to a single cumulative value by applying a function repeatedly.

Syntax: reduce(function, iterable)

Example:

python
from functools import reduce

numbers = [1, 2, 3, 4, 5]
sum_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_numbers)  # Output: 15
# 3. filter()
The filter() function applies a given function to each item in an iterable and returns an iterator with only the items for which the function returns True.

Purpose: To filter elements of an iterable based on a condition.

Syntax: filter(function, iterable)

Example:

python
numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4]
